```python
import pfun_cma_model as pfuncma
from pfun_cma_model import read_sample_data
import pandas as pd

df: pd.DataFrame = pd.DataFrame(read_sample_data(convert2json=False))
df
```

# Plot the raw sample data

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
df_plt = df.loc[df.index <= df.index[10]]
df_plt.set_index('ts_local', inplace=True)
df_plt = df_plt.sort_index()
sns.lineplot(df_plt, x='ts_local', y='sg')
```

# Run the CMA Model

```python
model = pfuncma.CMASleepWakeModel()

soln = model.run()
soln
```

## Plot the CMA Model Results

```python
from pfun_cma_model.engine.cma_plot import CMAPlotConfig

pfun_plot = CMAPlotConfig()

pfun_plot.plot_model_results(df=df, soln=soln)
```
